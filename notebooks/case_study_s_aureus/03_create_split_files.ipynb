### Measurement file

```python
import os
import pandas as pd


measurementsDf = pd.read_csv(os.environ['EHR_INT_ANALYSIS_BASE'] + '/data/case_study_1/data_matrix_final.csv')
measurementsDf
```

### Genomics file

```python
import os
import pandas as pd


genomicsDf = pd.read_csv(os.environ['EHR_INT_ANALYSIS_BASE'] + '/data/case_study_1/genomics/data_matrix_genomics.csv')
genomicsDf
```

```python
genomicsDf.patient_id = genomicsDf.patient_id.apply(lambda x: x[1:]).astype('int64')
genomicsDf
```

### LOS file

```python
import os
import pandas as pd


losDf = pd.read_csv(os.environ['EHR_INT_ANALYSIS_BASE'] + '/data/case_study_1/los_labels.csv')
losDf
```

```python
losDf = losDf.groupby(['person_id']).first().reset_index()
losDf
```

## Get common IDs

```python
mergedDf = demoDf.merge(
    measurementsDf['person_id'],
    how='inner',
    left_on='patient_id',
    right_on='person_id'
).drop_duplicates().drop(
    columns=['person_id']
).merge(
    genomicsDf[['patient_id']],
    how='inner',
    on='patient_id'
).drop(
    columns=['gender', 'birth_date']
)
mergedDf
```

## Save splits

### Measurements

```python
from pathlib import Path

from sklearn.model_selection import train_test_split


for wb, wa in [(365, 0), (365, 1), (365, 2), (365, 3)]:
    
    measurementsMergedDf = measurementsDf.merge(
        mergedDf,
        how='inner',
        left_on='person_id',
        right_on='patient_id'
    ).drop(
        columns=['patient_id']
    )
    
    measurementsMergedDf['measurement_date'] = pd.to_datetime(measurementsMergedDf['measurement_date'], format='%Y/%m/%d')
    measurementsMergedDf['admittime'] = pd.to_datetime(measurementsMergedDf['admittime'], format='%Y/%m/%d')

    measurementsMergedDf = measurementsMergedDf[(measurementsMergedDf['measurement_date'] >= (measurementsMergedDf['admittime'] - pd.Timedelta(days=wb))) & (measurementsMergedDf['measurement_date'] <= (measurementsMergedDf['admittime'] + pd.Timedelta(days=wa)))]
    measurementsMergedDf = measurementsMergedDf[measurementsMergedDf.person_id.isin(measurementsMergedDf[measurementsMergedDf.measurement_date == measurementsMergedDf.admittime].person_id.unique())]

    dropColumns = ['los_gt_7_days', 'los_gt_14_days', 'los_gt_30_days', 'los_gt_60_days', 'los_gt_90_days']
    filteredColumns = measurementsMergedDf.filter(dropColumns)
    measurementsMergedDf = measurementsMergedDf.drop(filteredColumns, axis=1)

    measurementsMergedDf = measurementsMergedDf.sort_values(by=['admittime'])
    idsDf = measurementsMergedDf[['person_id']].drop_duplicates()
    trainIdsDf, tempIdsDf = train_test_split(idsDf, train_size=0.85, shuffle=False, random_state=None)
    trainDataDf = measurementsMergedDf.merge(trainIdsDf, how='inner', on=['person_id'])
    validIdsDf, testIdsDf = train_test_split(tempIdsDf, test_size=0.66, shuffle=False, random_state=None)
    testDataDf = measurementsMergedDf.merge(testIdsDf, how='inner', on=['person_id'])
    validDataDf = measurementsMergedDf.merge(validIdsDf, how='inner', on=['person_id'])
    evaluateDataDf = measurementsMergedDf.merge(tempIdsDf, how='inner', on=['person_id'])
    splitsDirPath = Path(os.environ['EHR_INT_ANALYSIS_BASE'] + '/data/case_study_1/wb_' + str(wb) + '_wa_' + str(wa) + '/splits')
    measurementsPath = Path(splitsDirPath, 'measurements')
    measurementsPath.mkdir(parents=True, exist_ok=True)
    trainDataDf.to_csv(Path(measurementsPath, 'normal_train.csv'), index=False, sep='\t')
    testDataDf.to_csv(Path(measurementsPath, 'normal_test.csv'), index=False, sep='\t')
    validDataDf.to_csv(Path(measurementsPath, 'normal_validate.csv'), index=False, sep='\t')
    evaluateDataDf.to_csv(Path(measurementsPath, 'normal_evaluate.csv'), index=False, sep='\t')
```

### Demographics

```python
from pathlib import Path


for wb, wa in [(365, 0), (365, 1), (365, 2), (365, 3)]:

    splitsDirPath = Path(os.environ['EHR_INT_ANALYSIS_BASE'] + '/data/case_study_1/wb_' + str(wb) + '_wa_' + str(wa) + '/splits')
    measurementsPath = Path(splitsDirPath, 'measurements')
    demographicsPath = Path(splitsDirPath, 'demographics')
    demographicsPath.mkdir(parents=True, exist_ok=True)

    trainIdsDf = pd.read_csv(Path(measurementsPath, 'normal_train.csv'), sep='\t')[['person_id']].drop_duplicates()
    testIdsDf = pd.read_csv(Path(measurementsPath, 'normal_test.csv'), sep='\t')[['person_id']].drop_duplicates()
    validIdsDf = pd.read_csv(Path(measurementsPath, 'normal_validate.csv'), sep='\t')[['person_id']].drop_duplicates()
    evaluateIdsDf = pd.read_csv(Path(measurementsPath, 'normal_evaluate.csv'), sep='\t')[['person_id']].drop_duplicates()

    trainDataDf = demoDf.merge(
        trainIdsDf[['person_id']],
        how='inner',
        left_on='patient_id',
        right_on='person_id'
    ).drop(
        columns=['person_id']
    )
    trainDataDf.to_csv(Path(demographicsPath, 'normal_train.csv'), index=False, sep='\t')

    testDataDf = demoDf.merge(
        testIdsDf[['person_id']],
        how='inner',
        left_on='patient_id',
        right_on='person_id'
    ).drop(
        columns=['person_id']
    )
    testDataDf.to_csv(Path(demographicsPath, 'normal_test.csv'), index=False, sep='\t')

    validateDataDf = demoDf.merge(
        validIdsDf[['person_id']],
        how='inner',
        left_on='patient_id',
        right_on='person_id'
    ).drop(
        columns=['person_id']
    )
    validateDataDf.to_csv(Path(demographicsPath, 'normal_validate.csv'), index=False, sep='\t')

    evaluateDataDf = demoDf.merge(
        evaluateIdsDf[['person_id']],
        how='inner',
        left_on='patient_id',
        right_on='person_id'
    ).drop(
        columns=['person_id']
    )
    evaluateDataDf.to_csv(Path(demographicsPath, 'normal_evaluate.csv'), index=False, sep='\t')
```

### Genomics

```python
from pathlib import Path


for wb, wa in [(365, 0), (365, 1), (365, 2), (365, 3)]:

    splitsDirPath = Path(os.environ['EHR_INT_ANALYSIS_BASE'] + '/data/case_study_1/wb_' + str(wb) + '_wa_' + str(wa) + '/splits')
    measurementsPath = Path(splitsDirPath, 'measurements')
    genomicsPath = Path(splitsDirPath, 'genomics')
    genomicsPath.mkdir(parents=True, exist_ok=True)

    trainIdsDf = pd.read_csv(Path(measurementsPath, 'normal_train.csv'), sep='\t')[['person_id']].drop_duplicates()
    testIdsDf = pd.read_csv(Path(measurementsPath, 'normal_test.csv'), sep='\t')[['person_id']].drop_duplicates()
    validIdsDf = pd.read_csv(Path(measurementsPath, 'normal_validate.csv'), sep='\t')[['person_id']].drop_duplicates()
    evaluateIdsDf = pd.read_csv(Path(measurementsPath, 'normal_evaluate.csv'), sep='\t')[['person_id']].drop_duplicates()

    trainDataDf = genomicsDf.merge(
        trainIdsDf[['person_id']],
        how='inner',
        left_on='patient_id',
        right_on='person_id'
    ).drop(
        columns=['person_id']
    )
    trainDataDf.to_csv(Path(genomicsPath, 'normal_train.csv'), index=False, sep='\t')

    testDataDf = genomicsDf.merge(
        testIdsDf[['person_id']],
        how='inner',
        left_on='patient_id',
        right_on='person_id'
    ).drop(
        columns=['person_id']
    )
    testDataDf.to_csv(Path(genomicsPath, 'normal_test.csv'), index=False, sep='\t')

    validateDataDf = genomicsDf.merge(
        validIdsDf[['person_id']],
        how='inner',
        left_on='patient_id',
        right_on='person_id'
    ).drop(
        columns=['person_id']
    )
    validateDataDf.to_csv(Path(genomicsPath, 'normal_validate.csv'), index=False, sep='\t')

    evaluateDataDf = genomicsDf.merge(
        evaluateIdsDf[['person_id']],
        how='inner',
        left_on='patient_id',
        right_on='person_id'
    ).drop(
        columns=['person_id']
    )
    evaluateDataDf.to_csv(Path(genomicsPath, 'normal_evaluate.csv'), index=False, sep='\t')
```

### Create labels

#### LOS

```python
from pathlib import Path


for wb, wa in [(365, 0), (365, 1), (365, 2), (365, 3)]:

    splitsDirPath = Path(os.environ['EHR_INT_ANALYSIS_BASE'] + '/data/case_study_1/wb_' + str(wb) + '_wa_' + str(wa) + '/splits')
    measurementsPath = Path(splitsDirPath, 'measurements')
    losPath = Path(splitsDirPath, 'los')
    losPath.mkdir(parents=True, exist_ok=True)

    trainIdsDf = pd.read_csv(Path(measurementsPath, 'normal_train.csv'), sep='\t')[['person_id']].drop_duplicates()
    testIdsDf = pd.read_csv(Path(measurementsPath, 'normal_test.csv'), sep='\t')[['person_id']].drop_duplicates()
    validIdsDf = pd.read_csv(Path(measurementsPath, 'normal_validate.csv'), sep='\t')[['person_id']].drop_duplicates()
    evaluateIdsDf = pd.read_csv(Path(measurementsPath, 'normal_evaluate.csv'), sep='\t')[['person_id']].drop_duplicates()

    trainDataDf = losDf.merge(
        trainIdsDf[['person_id']],
        how='inner',
        on='person_id'
    )
    
    for los in [7, 14, 30, 60, 90]:
        trainDataDf[['person_id', 'los_gt_' + str(los) + '_days']]\
            .to_csv(Path(losPath, 'los_gt_' + str(los) + '_days_normal_train.csv'), index=False, sep='\t')

    testDataDf = losDf.merge(
        testIdsDf[['person_id']],
        how='inner',
        on='person_id'
    )

    for los in [7, 14, 30, 60, 90]:
        testDataDf[['person_id', 'los_gt_' + str(los) + '_days']]\
            .to_csv(Path(losPath, 'los_gt_' + str(los) + '_days_normal_test.csv'), index=False, sep='\t')

    validateDataDf = losDf.merge(
        validIdsDf[['person_id']],
        how='inner',
        on='person_id'
    )

    for los in [7, 14, 30, 60, 90]:
        validateDataDf[['person_id', 'los_gt_' + str(los) + '_days']]\
            .to_csv(Path(losPath, 'los_gt_' + str(los) + '_days_normal_validate.csv'), index=False, sep='\t')

    evaluateDataDf = losDf.merge(
        evaluateIdsDf[['person_id']],
        how='inner',
        on='person_id'
    )

    for los in [7, 14, 30, 60, 90]:
        evaluateDataDf[['person_id', 'los_gt_' + str(los) + '_days']]\
            .to_csv(Path(losPath, 'los_gt_' + str(los) + '_days_normal_evaluate.csv'), index=False, sep='\t')
```
