# Build Models

## Create data matrix

### Train Data

#### Read train file in tsv format

```python
import pandas as pd


casestudy1DirPath = '/home/vmadmin/workspace/EHR-Int-Analysis/data/case_study_1'

trainDataDf = pd.read_csv(casestudy1DirPath + '''/wb_365_wa_3/splits/measurements/normal_train.csv''', sep='\t')
trainDataDf
```

#### Read mortality labels file

```python
import pandas as pd


casestudy1DirPath = '/home/vmadmin/workspace/EHR-Int-Analysis/data/case_study_1'

trainLabelsDf = pd.read_csv(casestudy1DirPath + '''/wb_365_wa_3/splits/labels/mortality/mortality_7_day_normal_train.csv''')
trainLabelsDf
```

#### Merge train data and labels

```python
trainFinalDf = trainDataDf.merge(
    trainLabelsDf,
    how='inner',
    left_on=['person_id'],
    right_on=['patient_id']
).drop(columns=['patient_id'])
trainFinalDf
```

### Evaluate Data

#### Read evaluate file in tsv format

```python
import pandas as pd


casestudy1DirPath = '/home/vmadmin/workspace/EHR-Int-Analysis/data/case_study_1'

evaluateDataDf = pd.read_csv(casestudy1DirPath + '''/wb_365_wa_3/splits/measurements/normal_evaluate.csv''', sep='\t')
evaluateDataDf
```

#### Read mortality labels file

```python
import pandas as pd


casestudy1DirPath = '/home/vmadmin/workspace/EHR-Int-Analysis/data/case_study_1'

testLabelsDf = pd.read_csv(casestudy1DirPath + '''/wb_365_wa_3/splits/labels/mortality/mortality_7_day_normal_test.csv''')
testLabelsDf
```

```python
import pandas as pd


casestudy1DirPath = '/home/vmadmin/workspace/EHR-Int-Analysis/data/case_study_1'

validateLabelsDf = pd.read_csv(casestudy1DirPath + '''/wb_365_wa_3/splits/labels/mortality/mortality_7_day_normal_validate.csv''')
validateLabelsDf
```

```python
evaluateLabelsDf = pd.concat([testLabelsDf, validateLabelsDf], ignore_index=True)
evaluateLabelsDf
```

### Merge evaluate data and labels

```python
evaluateFinalDf = evaluateDataDf.merge(
    evaluateLabelsDf,
    how='inner',
    left_on=['person_id'],
    right_on=['patient_id']
).drop(columns=['patient_id']).drop_duplicates()
evaluateFinalDf
```

### Save train file in csv format

```python
casestudy1DirPath = '/home/vmadmin/workspace/EHR-Int-Analysis/data/case_study_1'
trainFinalDf.to_csv(casestudy1DirPath + '''/wb_365_wa_3/splits/measurements/comma_separated/normal_train.csv''', index=False)
```

### Save evaluate file in csv format

```python
casestudy1DirPath = '/home/vmadmin/workspace/EHR-Int-Analysis/data/case_study_1'
evaluateFinalDf.to_csv(casestudy1DirPath + '''/wb_365_wa_3/splits/measurements/comma_separated/normal_evaluate.csv''', index=False)
```

## Build and save the model

```python
import os


casestudy1DirPath = '/home/vmadmin/workspace/EHR-Int-Analysis/data/case_study_1'

for mortalityDays in [7]:

    print('--------------------------------------------------------------------------------------------------------------------')
    print('losDays: ', mortalityDays)
    print('--------------------------------------------------------------------------------------------------------------------')

    os.system(
        'cd ' + os.environ['EHR_ML_BASE'] + ';'
        +
        '''.venv/bin/python -m ehrml.ensemble.Build ''' + casestudy1DirPath + '''/wb_365_wa_3/splits/measurements/comma_separated/normal_train.csv -tc "death_''' + str(mortalityDays) + '''_day" -ic "person_id" -mdc "measurement_date" -adc "admittime" -wb 365 -wa 3 -sp ''' + casestudy1DirPath + '''/models/wb_365_wa_3_death_''' + str(mortalityDays) + '''_day.pkl;'''
        )
```

## Predict using the above model

```python
import os


casestudy1DirPath = '/home/vmadmin/workspace/EHR-Int-Analysis/data/case_study_1'

for mortalityDays in [7]:

    print('--------------------------------------------------------------------------------------------------------------------')
    print('losDays: ', mortalityDays)
    print('--------------------------------------------------------------------------------------------------------------------')

    os.system(
        'cd ' + os.environ['EHR_ML_BASE'] + ';'
        +
        '''.venv/bin/python -m ehrml.ensemble.Predict ''' + casestudy1DirPath + '''/wb_365_wa_3/splits/measurements/comma_separated/normal_evaluate.csv -tc "death_''' + str(mortalityDays) + '''_day" -ic "person_id" -mdc "measurement_date" -adc "admittime" -wb 365 -wa 3 -mp ''' + casestudy1DirPath + '''/models/wb_365_wa_3_death_''' + str(mortalityDays) + '''_day.pkl -sp ''' + casestudy1DirPath + '''/predictions/wb_365_wa_3_death_''' + str(mortalityDays) + '''_day.csv;'''
        )
```
