# Merge Genomic Files

## Merge individual genomic files

```python
import os

import pandas as pd

genomicDataDir = '/home/vmadmin/workspace/EHR-Int-Analysis/data/case_study_1/genomics/'

i = 0
dfs = []

for file in os.listdir(genomicDataDir):
    if file.endswith('.csv'):
        df = pd.read_csv(genomicDataDir + file)
        df = df.dropna()
        dfs.append(df[['patient_id', 'tube_id', 'token']])

print(len(dfs))
```

```python
mergedDf = pd.concat(dfs)
mergedDf
```

## Group the tokens for patient and tube ids

```python
groupedDf = mergedDf.groupby(['patient_id', 'tube_id']).agg({i:lambda x: ', '.join(list(x)) for i in mergedDf.columns[2:]}).reset_index()
groupedDf
```

## Save the grouped file

```python
genomicDataDir = '/home/vmadmin/workspace/EHR-Int-Analysis/data/case_study_1/genomics/'

groupedDf.to_csv(genomicDataDir + 'data_matrix_genomics.csv', index=False)
```

## Merge individual demographic files

```python
import os

import pandas as pd

demographicsDataDir = '/home/vmadmin/workspace/EHR-Int-Analysis/data/case_study_1/ehr_demographics/'

dfs = []

for file in os.listdir(demographicsDataDir):
    if file.endswith('.csv'):
        df = pd.read_csv(demographicsDataDir + file)
        dfs.append(df[['patient_id', 'gender', 'birth_date']])

print(len(dfs))
```

```python
mergedDf = pd.concat(dfs)
mergedDf
```

## Group the tokens for patient and tube ids

```python
groupedDf = mergedDf.groupby(['patient_id']).last().reset_index()
groupedDf
```

```python
genomicDataDir = '/home/vmadmin/workspace/EHR-Int-Analysis/data/case_study_1/'

groupedDf.to_csv(genomicDataDir + 'data_matrix_demographics.csv', index=False)
```
